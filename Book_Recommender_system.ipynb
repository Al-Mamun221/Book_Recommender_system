```python
import gradio as gr
import pickle
import pandas as pd

# Load data
model = pickle.load(open('artifacts/model.pkl', 'rb'))
book_names = pickle.load(open('artifacts/book_names.pkl', 'rb'))
book_pivot = pickle.load(open('artifacts/book_pivot.pkl', 'rb'))
books = pd.read_csv('BX-Books.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
books = books[['Book-Title', 'Image-URL-M']]

# Ensure book_names are a set for quick lookup
valid_books = set(book_names)

# Safe slicing for top 50 rated and trending books - only if they exist in book_names
top_50_rated = [book for book in book_names if book in valid_books][:50]
top_50_trending = [book for book in book_names if book in valid_books][50:100]

def recommend_with_images(book_name):
    book_id = list(book_names).index(book_name)
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    recommended_books = []
    for i in suggestions[0][1:]:
        title = book_names[i]
        image_url = books[books['Book-Title'] == title]['Image-URL-M'].values
        recommended_books.append((title, image_url[0] if len(image_url) > 0 else "https://via.placeholder.com/150"))
    return recommended_books

def display_recommendations(book_name):
    recommendations = recommend_with_images(book_name)
    html_content = """
    <style>
      body { margin: 0; font-family: 'Segoe UI', sans-serif; background-color: #0f172a; color: #aef1e2; }
      .navbar { background-color: #134e4a; padding: 15px; text-align: center; font-size: 24px; font-weight: bold; }
      .header {
        padding: 25px;
        text-align: center;
        font-size: 28px;
        font-weight: 900;
        color: #4ade80;
        background: linear-gradient(90deg, #134e4a 0%, #22c55e 100%);
        box-shadow: 0 4px 15px rgba(34, 197, 94, 0.6);
        border-radius: 12px;
        margin: 20px 0;
        font-family: 'Poppins', sans-serif;
        letter-spacing: 1.2px;
        text-transform: uppercase;
      }
      .grid-container {
        display: grid;
        grid-template-columns: repeat(auto-fit, minmax(150px, 1fr));
        gap: 20px;
        padding: 20px;
      }
      .book-item {
        text-align: center;
        background: #1e293b;
        border-radius: 8px;
        padding: 10px;
        transition: transform 0.3s;
      }
      .book-item:hover { transform: scale(1.05); box-shadow: 0 8px 16px rgba(255,255,255,0.1); }
      .book-image {
        width: 120px;
        height: 180px;
        object-fit: cover;
        border-radius: 5px;
      }
      .book-title {
        margin-top: 10px;
        font-size: 14px;
        color: #facc15;
      }
      .footer {
        text-align: center;
        padding: 15px;
        font-size: 14px;
        background-color: #134e4a;
        color: #aef1e2;
        margin-top: 30px;
      }
    </style>
    <div class="navbar"> BookWise</div>
    <div class="header">Recommended Books</div>
    <div class="grid-container">
    """

    for title, url in recommendations:
        html_content += f"""
        <div class="book-item">
          <img src="{url}" alt="{title}" class="book-image"/>
          <div class="book-title">{title}</div>
        </div>
        """
    html_content += "</div><div class='footer'> 2025 BookWise - All Rights Reserved</div>"
    return html_content

# Update dropdown based on filter option
def update_book_dropdown(filter_option):
    if filter_option == "Top 50 Rated":
        return gr.Dropdown.update(choices=top_50_rated, value=top_50_rated[0])
    elif filter_option == "Top 50 Trending":
        return gr.Dropdown.update(choices=top_50_trending, value=top_50_trending[0])
    else:
        return gr.Dropdown.update(choices=list(book_names), value=book_names[0])

# Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("## Book Recommendation System Using Machine Learning")
    gr.Markdown("Select a filter, then a book from the dropdown and click the button to see recommendations with images.")

    filter_dropdown = gr.Dropdown(choices=["All Books", "Top 50 Rated", "Top 50 Trending"], value="All Books", label="Filter Books By")
    book_dropdown = gr.Dropdown(choices=list(book_names), label="Select a Book")

    with gr.Row():
        recommend_btn = gr.Button("Show Recommendations")
        clear_btn = gr.Button("Clear")
    output = gr.HTML()

    filter_dropdown.change(update_book_dropdown, inputs=filter_dropdown, outputs=book_dropdown)
    recommend_btn.click(display_recommendations, inputs=book_dropdown, outputs=output)
    clear_btn.click(lambda: "", outputs=output)

demo.launch()
```
